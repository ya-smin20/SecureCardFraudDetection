```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
data_path = '../data/raw/creditcard.csv'  
df = pd.read_csv(data_path)
print(df.head())
```

```python
print(df.info())
```

# Summary statistics of the dataset

```python
print(df.describe())
```

# Checking for Missing Values

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
plt.title("Missing values heatmap")
plt.show()
```

```python
if 'Class' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Class', data=df)
    plt.title("Class Distribution")
    plt.show()
```

```python

if 'Class' in df.columns:
    class_counts = df['Class'].value_counts()
    print(class_counts)

    
    print(f"Class 0 (Non-Fraudulent): {class_counts[0]}")
    print(f"Class 1 (Fraudulent): {class_counts[1]}")
```

# Univariate Analysis

```python
numerical_features = df.select_dtypes(include=[np.number]).columns
df[numerical_features].hist(figsize=(16, 12))
plt.subplots_adjust(hspace=0.5, wspace=0.3) 
plt.suptitle("Distribution of Numerical Features", size=16)
plt.show()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(16, 12))
df[numerical_features].boxplot()
plt.title("Boxplot of Numerical Features")
plt.xticks(rotation=90)
plt.show()
```
